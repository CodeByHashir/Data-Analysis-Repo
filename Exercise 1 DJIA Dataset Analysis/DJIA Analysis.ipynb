### Setting up Data loading and import libraries

````python
# importing libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')

# Loading local dataset using pandas
Djia_data = pd.read_csv('djia.csv')
````

### Task 1: Summary Statistics

Write the code needed to calculate and print out:
1. The mean of the DJIA.
2. The median of the DJIA.
3. The standard deviation of the DJIA.
4. The minimum value of the DJIA.
5. The maximum value of the DJIA.


````python
# Calculating statistic analysis including mean, median, standard deviation, min and max
stats = {
    "Mean": Djia_data['Price'].mean(),
    "Median": Djia_data['Price'].median(),
    "Standard Dev": Djia_data['Price'].std(),
    "Min": Djia_data['Price'].min(),
    "Max": Djia_data['Price'].max()
}

for stat,  values in stats.items():
    print(f"{stat}: {values:.2f}")
````

# Statiscal Insights 
1. The summary statistics show a mean of about 290.81 and a median of about 172.27, indicating a fairly large gap between the two. This gap often suggests skewed distribution, like the presence of outliers in the dataset. 
2. The standard deviation of 256.06 is also quite large relative to the mean, which indicates high variability in the data. 
3. Meanwhile, the MIN (46.85) and MAX (985.93) show a wide range of values. 

## conclusion:
Overall, these figures imply that the data might not be normally distributed and could contain extreme values driving the average upward. 

### Task 2: Box Plot

Write the code that will create a box plot for your data, using the boxplot function
built into the â€˜DataFrameâ€™ variable Djia_data`. Ensure the plot has,
1. A meaningful title.
2. A meaningful x-axis label.



````python
# Creating a Box or whisker plot
plt.figure(figsize=(10, 7))
Djia_data.boxplot(column='Price', grid=False, patch_artist=True, boxprops=dict(facecolor='blue'))

# Drawing plot
plt.title("Distribution of DJIA Prices (1914-1968)", fontweight='bold')
plt.xlabel("DJIA Price (USD)", fontstyle='italic')
plt.tight_layout()
plt.show()
````

This box plot shows the distribution of DJIA prices from 1914â€“1968. Notice that several outliers lie well above most of the data points, indicating a skewed distribution. The median is closer to the lower part of the box, reflecting a concentration of values on the lower end. However, outliers near 1000 push the overall range upward. This pattern suggests that while many data points cluster around moderate price levels, a few significantly higher DJIA prices create a wide spread in the data.

````python
# Creating violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(y='Price', data=Djia_data, inner='quartile', color='skyblue')

# Customizing plot
plt.title("Density Distribution of DJIA Prices", fontweight='bold')
plt.ylabel("DJIA Price (USD)", fontstyle='italic')
plt.tight_layout()
plt.show()
````

## Task 4: Line Chart

Write the code that will create a line chart for your data using the plot function built
into the **DataFrame** variable  ```djia_data``. Ensure the plot has,
1. A meaningful title.
2. Meaningful x-axis and y-axis labels


````python
# Convert 'Date' to datetime and sort (critical step)
Djia_data['Date'] = pd.to_datetime(Djia_data['Date'])
Djia_data.sort_values('Date', inplace=True)

# Create line chart
plt.figure(figsize=(12, 6))
plt.plot(Djia_data['Date'], Djia_data['Price'], color='navy', linewidth=1)

# Customize plot
plt.title("Historical Trend of DJIA Prices year 1914-1968", fontweight='bold')
plt.xlabel("Year", fontstyle='italic')
plt.ylabel("Price (USD)", fontstyle='italic')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

## Historical Trend of DJIA Prices (1914â€“1968)

This line plot illustrates the historical trend of DJIA prices from 1914 to 1968.  

### Key Observations:  
- The **early years (1914â€“1929)** show a steady increase, followed by a **sharp drop** during the Great Depression (1929â€“1932).  
- Between **1930 and 1950**, prices remained relatively volatile but did not experience sustained growth.  
- **Post-1950**, there is a clear upward trend, with significant growth in the **1960s**, reflecting economic expansion and market confidence.  
- The **steep rise in the late 1950s and 1960s** indicates a period of strong stock market performance.  

This visualization highlights the long-term growth of the DJIA despite short-term economic downturns, demonstrating the resilience and expansion of the stock market over time.


## Task 5: Line of Best Fit

Write the code that will create a line of best fit chart for your data using the numpy
â€˜polyfitâ€™ and â€˜poly1dâ€™ functions. There are a few steps to complete here:
1. Create data to represent the x-axis since you canâ€™t fit a line of best fit directly to
date time data. You can use code such as â€˜x = list(range(0, 649))â€™ to create the x-axis
data you need.
2. Calculate the coefficients using â€˜polyfitâ€™.
3. Create the line of best fit by passing the coefficients calculated above to poly1d.
4. Plot the line of best fit.
A little research will help you tackle this challenge. Be sure your plot has:
â€¢ A meaningful title.
â€¢ Meaningful x-axis and y-axis labels

````python
# Generate numeric x-axis 
x = np.arange(len(Djia_data)) # i use arange method which is more efficient with numpy and it is equivalent to x=list(range(0, 649))
y = Djia_data['Price']
````

````python
# Calculate ploynomial fit degree 1 for linear
coefficients = np.polyfit(x, y, 1)
poly = np.poly1d(coefficients)
trend_line = poly(x)
````

````python
# Plot data and trend line
plt.figure(figsize=(12, 6))
plt.scatter(x, y, s=5, color='gray', label='Actual Data')
plt.plot(x, trend_line, color='red', linewidth=2, label='Linear Trend')

# Customize plot
plt.title("DJIA Price Trend with Line of Best Fit", fontweight='bold')
plt.xlabel("Months Since December 1914", fontstyle='italic')
plt.ylabel("Price (USD)", fontstyle='italic')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print trend equation (professional insight)
print(f"Trend Equation: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}")
````
